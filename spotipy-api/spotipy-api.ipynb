```python
import os
import spotipy

from spotipy.oauth2 import SpotifyClientCredentials
from dotenv import load_dotenv

load_dotenv()
```

```python
class Spotipy:

    def __init__(self, show_id):
        self.sp = spotipy.Spotify(
            auth_manager = SpotifyClientCredentials(
                client_id = os.environ.get("SPOTIPY_CLIENT_ID"), 
                client_secret = os.environ.get("SPOTIPY_CLIENT_SECRET")))
        self.show_id = show_id

    def get_episodes(self):
        offset = 0
        episodes = []
        
        while True:
            episode_batch = self.sp.show_episodes(
                show_id=self.show_id, 
                limit=50, 
                offset=offset, 
                market='us'
            )['items']
            
            if not episode_batch:
                break

            for episode in episode_batch:
                episodes.append({
                    'spotify_id': episode['id'],
                    'name': episode['name'],
                    'description': episode['html_description'],
                    'image': episode['images'][1]['url'],
                    'spotify_url': episode['external_urls']['spotify'],
                })
                    
            offset += 50
        
        return episodes
```

```python
show_id = '0xOdWuktBQKWCv2mlVjizn'

sp = Spotipy(show_id=show_id)

episodes = sp.get_episodes()

len(episodes)
```
